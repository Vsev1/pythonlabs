Отримати квадратну матрицю порядку $n$

![Image](http://i.piccy.info/i9/8ecdafefb6d7debd9ca72284d5aca7f6/1620854983/5867/1427906/80736yzobrazhenye_2021_05_13_002959.png)

### Код

```python
n = int(input("n = "))
a = [[j + i if j + i <= n else 0 for j in range(1, n + 1)] for i in range(n)]
print(*a, sep="\n")
```
