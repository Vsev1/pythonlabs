Даны целые числа ($a_1$ ..., $a_n$, $b_1$ ..., $b_n$).
Верно ли, что эти две последовательности отличаются не более
чем порядком следования членов?

```python
import collections
import random

n = int(input("Введіть n: "))
nums = [random.randint(-10, 10) for _ in range(n)]
nums2 = [random.randint(-10, 10) for _ in range(n)]
# nums = [1, 2, 3, 4, 5]
# nums2 = [3, 5, 2, 1, 4, 3]
print(nums, nums2)
print("Не відрізняються" if collections.Counter(nums) == collections.Counter(nums2) else "Відрізняються")
```
