### Завдання 695

Таблица футбольного чемпионата, в котором участвовало $n$
команд (см. задачу 413), задана своей верхней
правой частью в виде последовательности чисел 0, 1 или 2
первые $n$ - 1 чисел последовательности относятся к первой
строке таблицы, следующие $n$ - 2 чисел — ко второй и т. д.
Построить таблицу целиком, т. е. получить соответствующую
квадратную матрицу порядка $n$ (элементы главной
диагонали заполняются нулями).

```python
n = int(input("Введіть n: "))
a = [[0 if j == i else j + i for j in range(n)] for i in range(n)]
print(*a, sep="\n")
```
