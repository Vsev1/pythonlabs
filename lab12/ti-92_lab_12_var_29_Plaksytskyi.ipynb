Даны натуральное число $n$, действительные числа $y_1$, ..., $y_n$. Найти:

г) $$(\sqrt{z_1} - z_1)^2 + ... + (\sqrt{z_n} - z_n)^2$$,
   где $z_i$ = \begin{cases}
   y_i &\text{при } 0 < y_i \leq 15,
   \\
   2.7 &\text{в противном случае}
   \end{cases}

```python
import random
from math import sqrt

n = int(input("Введіть n: "))
nums = [random.randint(-10, 10) for i in range(n)]
print(nums)
print(sum((sqrt(i) - i) ** 2 if 0 < i <= 15 else 2.7 for i in nums))
```
