### Завдання 116д

Даны натуральное число $n$, действительное число $x$.
Вычислить:
д) $$ \prod_{k=1}^n (\frac{k}{k+1}-(\cos{|x|})^k) $$

```python
from math import cos, prod

n = int(input("Введіть n: "))
x = float(input("Введіть x: "))

print(prod(k / (k + 1) - cos(abs(x)) ** k for k in range(1, n + 1)))
```
