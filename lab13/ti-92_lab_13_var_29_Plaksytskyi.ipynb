Даны действительные числа $a_1$, ..., $a_{30}$. Получить:
а) max($a_1$ + $a_{30}$, $a_2$ + $a_{29}$, ..., $a_{15}$ + $a_{16}$)
б) min($a_1$ * $a_{16}$, $a_2$ * $a_{17}$, ..., $a_{15}$ * $a_{30}$)

```python
import random

n = int(input("Введіть n: "))
nums = [random.randint(-10, 10) for i in range(n)]
print(nums)
print(max(i + j for i, j in zip(nums[:n // 2 + 1], nums[n // 2 + 1::-1])))
print(min(i * j for i, j in zip(nums[:n // 2 + 1], nums[n // 2 + 1:])))
```
