### Завдання 223д

Даны целые числа $a_1$, $a_2$, ... Известно, что $a_1$ > 0
и что среди $a_2$, $a_3$, ... есть хотя бы одно отрицательное
число. Пусть $a_1$, ... , $a_n$ — члены данной последовательности, предшествующие первому отрицательному члену
($n$ заранее неизвестно). Получить:
д) max($a_1$, $a_1$$a_2$, ..., $a_1$$a_2$...$a_n$);

```python
import math
from itertools import takewhile

nums = [7, 6, 5, 8, 2, 3, -2, -5, 3, -5, -8]
print(nums)
print(math.prod(n for n in takewhile(lambda num: num >= 0, nums)))
```
