Дано натуральное число $n$. Получить последовательность $b_1$ ..., $b_n$, где при $i$ = 1, 2, . . . , $n$ значение $b_i$ равно:  
д) $$2^i + 3^{i+1}$$

```python
n = int(input("Введіть n: "))
print([2 ** i + 3 ** (i + 1) for i in range(1,n + 1)])
```
